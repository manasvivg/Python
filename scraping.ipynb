```python
import requests
import urllib2
import bs4
from pattern import web
from pandas import DataFrame, Series
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
```

```python
url = 'http://www.imdb.com/chart/top'
r = requests.get(url)
```

```python

#http://www.imdb.com/search/title?at=0&release_date=1950,2012&sort=num_votes&title_type=feature
url = 'http://www.imdb.com/search/title?count=250&release_date=1950,2012&title_type=feature&sort=num_votes'
#params = dict(sort='num_votes,desc', start=1, title_type='feature', year=1950,2012)
r1 = requests.get(url+'&start=1')
r2 = requests.get(url+'&start=101')
r3 = requests.get(url+'&start=201')
r4 = requests.get(url+'&start=301')
r5 = requests.get(url+'&start=401')
r6 = requests.get(url+'&start=501')
r7 = requests.get(url+'&start=601')
r8 = requests.get(url+'&start=701')
r9 = requests.get(url+'&start=801')
r10 = requests.get(url+'&start=901')

#r = [r1.text,r2.text,r3.text]
#Data scraped at 3:37PM 2014-11-11
```

```python
r = [r1.text, r2.text, r3.text, r4.text,r5.text,r6.text,r7.text,r8.text,r9.text,r10.text]
i = 1
for text in r:
    dom =  web.Element(text)
    title = []
    runtime = []
    year = []
    rating = []
    genres1 = []
    votes = []

    for movie in dom.by_tag('td.title'):
        title.append(movie.by_tag('a')[0].content)
        runtime.append(movie.by_tag('span.runtime')[0].content)
        year.append(movie.by_tag('span.year_type')[0].content)
        rating.append(movie.by_tag('span.value')[0].content)
        genres =  movie.by_tag('span.genre')[0]
        genres1.append([a.content for a in genres.by_tag('a')])
    
    genres2 = ['|'.join(x) for x in genres1]  
    
    for aux in dom.by_tag('td.sort_col'):
        votes.append(aux.content.replace(',',''))
    
    df = DataFrame({'movie' : title, 'year' : year, 'runtime' : runtime, 
                        'genres' : genres2, 'rating' : rating,'votes' : votes})
    if i == 1:
        movies = df.copy()
    if i > 1:
        movies = pd.concat([movies,df])
    i+=1
    
movies.index = range(len(movies))
```

```python
a = movies.genres.str.get_dummies(sep='|')
a.columns = [ 'genre_' + x for x in list(a.columns)]
movies = movies.merge(a,left_index=True,right_index=True)
```

```python
movies.rating = movies.rating.astype(float)
movies[['votes']] = movies[['votes']].astype(int)
movies.year = pd.Series(movies.year).str.replace('(', '').str.replace(')', '').astype(int)
movies.runtime = pd.Series(movies.runtime).str.replace(' mins.', '').astype(int)
```

### Exploratory Analysis

```python
movies1 = pd.melt(movies,id_vars=['movie','rating','genres','votes','runtime','year'])
```

```python
movies1 = movies1[movies1['value'] == 1]
movies1.variable = Series(movies1.variable).str.replace('genre_','')
```
