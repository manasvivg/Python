```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from pandas import DataFrame, Series
import pandas as pd
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
from numpy.random import randn
plt.plot(randn(50).cumsum(),'k--')
_ = ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
ax2.plot((0,35),(0,35),'r--')
ax3.plot((0,50),(10,10),'b-')
```

```python
ax3.plot((0,50),(0,0),'b-')
```

```python
colors = ['k','r','b','g']
z=0
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) 
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color=colors[z], alpha=0.5)
        z += 1
plt.subplots_adjust(wspace=.05, hspace=.05)
```

```python

plt.plot(randn(30).cumsum(), 'b-o')
```

```python
colors = ['k','r','b','g']
fig, axes = plt.subplots(2,2)
axes1 = axes.reshape(-1)
for i in range(len(colors)):
    data = randn(30).cumsum()
    axes1[i].plot(data, color=colors[i],linestyle='-', label='Default',alpha=0.65)
    axes1[i].plot(data, color=colors[i], linestyle=':',drawstyle='steps-post', label='steps')
    axes1[i].legend(loc='best')
    axes1[i].set_xticks([0,10,20,30])
    axes1[i].set_yticks([-10,-5,0,5,10,15])
```

```python
data = randn(30)
plt.plot(data.cumsum(), 'b-o')
#plt.text(20,data.mean(),'This is \na test',fontsize=14)
plt.arrow(20,data.mean(),2,0,color="r",width=.1,head_width =.5)
```

```python
df = DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'],  
               index=np.arange(0, 100, 10))
```

```python
tips = pd.read_csv('ch08/tips.csv')
party_counts = pd.crosstab(tips.day, tips.size)
party_pcts = party_counts.div(party_counts.sum(1).astype(float), axis=0)
party_pcts.plot(kind='bar', stacked=True,alpha=0.7)
```

```python
aux = party_counts.stack().unstack(0)
party_pcts1= aux.div(aux.sum(1).astype(float),axis=0)
party_pcts1.plot(kind='bar',stacked=True)
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips['tip_pct'].hist(bins=50,normed=True)
tips['tip_pct'].plot(kind='kde',color='r')
```
