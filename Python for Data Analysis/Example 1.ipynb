```python
time_zones = [x['tz'] for x in records if 'tz' in x]
```

```python
len(time_zones)
```

```python
#Base python way
def get_counts(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x] +=1
        else:
            counts[x] = 1
    return counts
counts = get_counts(time_zones)
counts.items()[1]
```

```python
#use defaultdict, a bit more efficient
from collections import defaultdict
def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] += 1
    return counts

counts2 = get_counts2(time_zones)
print counts2['Europe/Madrid']
```

```python
#sort normal way
value_key_pairs = [(x[1],x[0]) for x in counts.items()]
value_key_pairs.sort()
print 'top 10 time zones:',value_key_pairs[-10:]
```

```python
#Better Way, with Counter function
from collections import Counter
counts3 = Counter(time_zones)
counts3.most_common(10)
```

```python
unames = ['user_id','gender','age','occupation','zip']
users = pd.read_table('/Users/mario/Documents/Python/Python for Data Analysis/users.dat.txt',sep="::",header=None,names=unames)
rnames = ['user_id','movie_id','rating','time_stamp']
ratings = pd.read_table('/Users/mario/Documents/Python/Python for Data Analysis/ratings.dat.txt',sep="::",header=None,names=rnames)
mnames = ['movie_id','title','genres']
movies = pd.read_table('/Users/mario/Documents/Python/Python for Data Analysis/movies.dat.txt',sep="::",header=None,names=mnames)
```

```python
#merge the data
data = pd.merge(ratings,users,how="inner",left_on="user_id",right_on="user_id")
data = pd.merge(data,movies,how="inner",left_on="movie_id",right_on="movie_id")
data.dtypes
```

```python
mean_ratings = data.pivot_table('rating',rows='title',cols='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
active_titles = ratings_by_title.index[ratings_by_title>=250]
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings[:10]
```

```python
top_female_ratings = mean_ratings.sort_index(by='F',ascending=False)
top_female_ratings[:5]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
sorted_by_diff = mean_ratings.sort_index(by='diff')
print sorted_by_diff[:10]
print sorted_by_diff[-10:]
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
rating_std_by_title = rating_std_by_title.ix[active_titles]
rating_std_by_title.order(ascending=False)[-10:]
```

##Baby Names Example Page 32

```python
names1880 = pd.read_csv('/Users/mario/Documents/Python/Python for Data Analysis/ch02/names/yob1880.txt',names=['name','sex','births'])
```

```python
names1880.groupby('sex').births.sum()
```

```python
years = range(1880,2011)
pieces=[]
columns = ['name','sex','births']

for year in years:
    path = '/Users/mario/Documents/Python/Python for Data Analysis/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path,names=columns)
    frame['year'] = year
    pieces.append(frame)
    
#concat all
names = pd.concat(pieces,ignore_index=True)
```

```python
%matplotlib inline
names.groupby('year').births.sum().plot()
```

```python
total_births = names.pivot_table('births',rows='year',cols='sex',aggfunc=sum)
total_births.plot(title="Total Births by Year and Sex")
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year','sex']).apply(add_prop)
```

```python
np.allclose(names.groupby(['year','sex']).prop.sum(),1)
```

```python
def get_top1000(group):
    return group.sort_index(by='births',ascending=False)[:1000]

grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(get_top1000)
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births',rows='year',cols='name',aggfunc=sum)
```

```python
subset = total_births[['John','Harry','Mario','Jordan','Mary','Marilyn','Zoe']]
```

```python
subset.plot(subplots=False, figsize=(12,10),grid=False,title="Number of Births per Year - Selected Names")
```

```python
table = top1000.pivot_table('prop',rows='year',cols='sex',aggfunc=sum)
table.plot(title="Proportion of Top 1000 Names" , yticks=np.linspace(0,1.1,12))
```

```python
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'
table = names.pivot_table('births',rows=last_letters,cols=['sex','year'],aggfunc=sum)
```

```python
subtable= table.reindex(columns=[1910,1960,2010],level='year')
subtable.head()
subtable.sum()
letter_prop = subtable / subtable.sum().astype(float)
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,1,figsize=(10,8))
letter_prop['M'].plot(kind='bar',rot=0,ax=axes[0],title='Male')
letter_prop['F'].plot(kind='bar',rot=0,ax=axes[1],title='Female',legend=False)
```
