```python
#Load Libraries to be used
from pandas import *
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#Read Data
f = open('log.csv','r')
log = pd.read_csv('log.csv')
year = Series(2000+(log['Date'].str[-2:]).astype(int))
log['year'] = year
log1 = log[log['year'] >= 2012]
```

```python
#calculate cummulative sum by year
#log1 = log1.reset_index()
#log1 = log1.set_index(["year"]) 
log1['cummulative'] = np.cumsum(log1.groupby(['year'])['Distance'])
log1.set_index(log1['year'])
```

```python
#do chart
%matplotlib inline

plt.plot(log1['cummulative'][log1['year']==2012], 'g', label="2012")
plt.plot(log1['cummulative'][log1['year']==2013], 'b', label="2013")
plt.plot(log1['cummulative'][log1['year']==2014], 'r', label="2014")
plt.legend(loc=4)
```

```python
#I want to convert the date to a date and then calculate days before race
log1['date'] = pd.to_datetime(log1['Date'])
#log1['days'] = Series().astype('timedelta64')

#add race day
log1.loc[log1.year==2014,'race_date'] = pd.to_datetime("10/11/2014")
log1.loc[log1.year==2013,'race_date'] = pd.to_datetime("11/17/2013")
log1.loc[log1.year==2012,'race_date'] = pd.to_datetime("10/13/2012")

#Now Calculate days before race
log1['days'] = (log1['date'] - log1['race_date'])
log1['days1'] = (log1['days'] / np.timedelta64(1, 'D')).astype(int)
```

```python
log2 =log1[['days','Distance','year','Date','date','days1','Duration']]
log2 = log2[(log2.days1 >= -90.0) & (log2.days1 <=0.0)]

#[((log1.days/np.timedelta64(1,'D')) >= -90) * ((log1.days/np.timedelta64(1,'D')) <= 0)]
log2.sort(['year','date'])
log2['cummulative'] = np.cumsum(log2.groupby(['year'])['Distance'])

aux = log2[['days1','cummulative','year']][(log1.year==2012)]
plt.plot(aux['days1'],aux['cummulative'], 'g', label="2012")
aux = log2[['days1','cummulative','year']][(log1.year==2013)]
plt.plot(aux['days1'],aux['cummulative'], 'b', label="2013")
aux = log2[['days1','cummulative','year']][(log1.year==2014)]
plt.plot(aux['days1'],aux['cummulative'], 'r', label="2014")
plt.xlabel('Days Before Race')
plt.ylabel('Cummulative Distance')
plt.legend(loc=4)
plt.title('Training Program for Half- Marathon')
plt.draw()
```

```python
#Analyze the pace
log2.index = range(len(log2)) 

from datetime import datetime 
format = '%H:%M:%S' 
log2['duration'] = [datetime.strptime(x, format) for x in log2['Duration']]
log2['minutes'] = [x-datetime.strptime("0:00:00", format) for x in log2['duration'] ]
log2['minutes'] = [x.total_seconds()/60 for x in log2['minutes']]
log2['pace'] = log2['minutes']/log2['Distance']

#import re
#sometimes the log has no hours, I have to add the '0:'
#for z in range(len(log2.index)): 
#    match = re.search(r'([0-9]+:[0-9]+)(?:.[0-9]+)',log2['Duration'][z]) 
#    if match: 
#        log2['duration'][z] = '00:' + match.group(1)
#    else:
#        log2['duration'][z] = log2['Duration'][z]
        

#times =[datetime.strptime(x, format) for x in log2['Duration']] 
#print times.astype('datetime64[M]')
```
