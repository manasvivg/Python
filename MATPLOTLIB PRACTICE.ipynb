```python
#load libraries
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#first step is to open a figure
fig = plt.figure(figsize=(20,10))
fig.gca()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(10),np.random.randn(10),'bx')
ax.set_ylim(1000,500)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figheight(6)
fig.set_figwidth(10)
ax1.scatter(np.random.randn(10),np.random.randn(10),color='b',marker='*')
ax2.scatter(np.random.randn(10),np.random.randn(10),color='r',marker='s')
fig.suptitle("title")
fig.gca()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_figheight(6)
fig.set_figwidth(10)
ax.plot([7,8,9,10,11],[50,48,52,55,57],'cs-',label="Stamford, CT")
ax.plot([7,8,9,10,11],[68,73,72,70,64],'go-',label="San Francisco, CA")
ax.plot([7,8,9,10,11],[32,32,32,32,32],'b:',label="_nolegend_")
ax.plot([7,8,9,10,11],[75,75,75,75,75],'r:',label="_nolegend_")
ax.set_ylim(30,80)
ax.legend()
ax.xaxis.set_ticks(range(7, 12,1))    
ax.xaxis.set_ticklabels(map(lambda x: 'Nov ' + str(x), range(7, 12,1)))
ax.set_title('Maximum Daytime Temperature',size=18)
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(122)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Philadelphia')
ax.plot([1, 2, 3, 4], [30, 23, 13, 4], label='Boston')
ax.set_title('A tale of two cities')
ax.legend()
major_formatter = plt.FormatStrFormatter('$%2.1f')
ax.xaxis.set_major_formatter(major_formatter)

t = np.linspace(0, 7, 25)
z = 2 * np.sin(t) + 5
ax = fig.add_subplot(121)
ax.scatter(t, z, label='Philadelphia')
ax.set_title("Observed Tide")
ax.legend()

```
